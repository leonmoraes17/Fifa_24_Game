#MACHINE LEARNING FOR CALCULATING "OVERALL" RATING 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split (X, y, test_size= 0.20)
```

#TESTINg PIPELINe

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('regression', LinearRegression())
])
pipe.fit(X_train, y_train)

y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import r2_score # goodness of fit of line : R2 score to be near 1 means the we can explain 75% of variance of dependent variable based on independent variable.
print( r2_score (y_test, y_pred))

df_predicted_overall = pd.DataFrame(
    {
        "Overall_24": y_test,
        "pred_overall_24": y_pred
    }
)

df_predicted_overall
```

```python
# We  are trying to find if there is statistically significant difference between the means(Value of player) of 2 groups( Midfielder, Attacker) of one category ( Player Positions) using Independent T-test
# Ho = There is no difference in value of player depending on his position ; H1 : There is difference in value of player depending on his position . P-value <= 0.05 ( 5%) which means that the data is likely to occur less than 5% of the time under the null hypothesis and hence we 
# reject the null hpyothesis( which means less than 5% of the time the null hyptoheseis is accepted)
df_24['player_position'] = df_24['club_position_24'].map({'CAM': 'Midfield', 'CM':'Midfield','LAM': 'Midfield', 'LM':'Midfield', 'LW': 'Midfield', 'RAM':'Midfield','RCM': 'Midfield', 'RM':'Midfield',
                                                       'RS': 'Midfield', 'RW':'Midfield','RES': 'Midfield', 'LCM':'Midfield','LS': 'Midfield', 'RB':'Defender',
                                                        'CB':'Defender', 'CDM': 'Defender','LB':'Defender', 'LCB': 'Defender','LDM':'Defender', 'RCB': 'Defender','RDM':'Defender', 'RWB': 'Defender',
                                                         'LWB':'Defender', 'CF': 'Forward','LF':'Forward', 'SUB': 'Forward', 'ST':'Forward', 'RF':'Forward' })
```

```python
#Independent 2 sample Ttest
# Null Hypothesis : There is no difference in average value of the player depending on the position he plays
import scipy.stats as stats

# Extract individual groups
# Forward = df_24[df_24["player_position"]== 'Forward']['height_cm_24']
# Midfield = df_24[df_24["player_position"]== 'Midfield']['height_cm_24']
# Defender = df_24[df_24["player_position"]== 'Defender']['height_cm_24']

Forward = df_24["value_eur_24"][df_24["player_position"]=="Forward"]
Midfield = df_24["value_eur_24"][df_24["player_position"]=="Midfield"]


#running Independent T-Test to check the Null Hypothesis 
stats.ttest_ind(Midfield, Forward, equal_var=False)

#We are rejecting the Null hyptothesis: Which means plays position influences the "Value" of the player. 
```

```python
forward_sample = Forward.sample(500)
midfield_sample = Midfield.sample(500)
stats.ttest_ind(midfield_sample, forward_sample, equal_var=False)
```
